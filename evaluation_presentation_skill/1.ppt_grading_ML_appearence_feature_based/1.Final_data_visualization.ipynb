```python
#data visualization
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
df=pd.read_csv("E:/3.dataset_150.csv")
X=df.iloc[:,:22]
y=df['output_grade']
yc = df.iloc[:,22:23].values
```

```python
# Show first two principal components without scaler

pca = PCA()
plt.figure(figsize=(8,6))
Xt = pca.fit_transform(X)
plot = plt.scatter(Xt[:,0], Xt[:,1], c=yc)
plt.legend(handles=plot.legend_elements()[0], labels=[1,2,3,4])
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("First two principal components")
plt.show()


```

```python
# Show first two principal components with scaler
pca = PCA()
pipe = Pipeline([('scaler', StandardScaler()), ('pca', pca)])
plt.figure(figsize=(8,6))
Xt = pipe.fit_transform(X)
plot = plt.scatter(Xt[:,0], Xt[:,1], c=df['output_grade'])
plt.legend(handles=plot.legend_elements()[0], labels=[1,2,3,4])
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("First two principal components after scaling")
plt.show()
```

```python
#Visualizaing 3D plot
fig, ax = plt.subplots()
ax = fig.add_subplot(projection='3d')
ax.scatter(Xt[:,1], Xt[:,2], c=yc)
plt.show()
```
