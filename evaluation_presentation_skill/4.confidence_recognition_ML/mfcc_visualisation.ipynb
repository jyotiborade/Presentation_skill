```python
import librosa
%matplotlib inline
import matplotlib.pyplot as plt
import librosa.display
low = 'C:\\Users\DELL\Desktop\speech\\low.wav'
x , sr = librosa.load(low)
high = 'C:\\Users\DELL\Desktop\speech\\high.wav'
y , sr = librosa.load(high)

g = librosa.feature.mfcc(x, sr=sr)
k = librosa.feature.mfcc(y, sr=sr)

fig, ax = plt.subplots(1,2, figsize = (8,4))

ax[0].set(title = '')
i = librosa.display.specshow(g,x_axis='time', ax=ax[0],cmap='coolwarm')





ax[1].set(title = '')
librosa.display.specshow(k,x_axis='time',ax=ax[1],cmap='coolwarm')

plt.colorbar(i)

```
