```python
import librosa
%matplotlib inline
import matplotlib.pyplot as plt
import librosa.display
import warnings
warnings.filterwarnings('ignore')
low = 'C:\\Users\DELL\Desktop\speech\\low.wav'
x , sr = librosa.load(low)
high = 'C:\\Users\DELL\Desktop\speech\\high.wav'
y , sr = librosa.load(high)

g = librosa.feature.chroma_stft(y=x, sr=sr)
k = librosa.feature.chroma_stft(y=y, sr=sr)

fig, ax = plt.subplots(1,2, figsize = (9,4))
ax[0].set(title = '')
i = librosa.display.specshow(g,x_axis='time', y_axis='chroma',ax=ax[0],cmap='coolwarm')

ax[1].set(title = '')
librosa.display.specshow(k,x_axis='time',y_axis='chroma', ax=ax[1],cmap='coolwarm')

plt.colorbar(i)
```
