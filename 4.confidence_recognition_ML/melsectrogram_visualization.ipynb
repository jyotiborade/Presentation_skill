```python
import librosa
low = 'C:\\Users\DELL\Desktop\speech\\low.wav'
x , sr = librosa.load(low)
high = 'C:\\Users\DELL\Desktop\speech\\high.wav'
y , sr = librosa.load(high)
```

```python
sr=22050 (sampling rate)
hop_length=512 (number of samples between frames)
n_fft=2048 (number of samples per frame in STFT-like analyses
```

```python
#display waveform
%matplotlib inline
import matplotlib.pyplot as plt
import librosa.display

x = librosa.feature.melspectrogram(x)
y = librosa.feature.melspectrogram(y)

# Convert amplitudes to dB
g = librosa.amplitude_to_db(x)
k = librosa.amplitude_to_db(y)
# Plot mel spectrograms
fig, ax = plt.subplots(1,2, figsize = (10,5))
ax[0].set(title = '')
i = librosa.display.specshow(g,x_axis='time',y_axis='mel', ax=ax[0],cmap='coolwarm')

ax[1].set(title = '')
librosa.display.specshow(k,x_axis='time', y_axis='mel',ax=ax[1],cmap='coolwarm')

plt.colorbar(i)
```
